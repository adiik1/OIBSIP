## ANALYSIS OF DATA

```python
numeric_rd = rd.select_dtypes(include=['number'])
corr= numeric_rd.corr()

sns.heatmap(corr, annot=True)
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.title('Unemployment In India')
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=rd)
plt.show()
```

```python
rd.columns
plt.title("Unemployment")
sns.histplot(x="Estimated Employed",hue="Region",data=rd)
plt.show()
```

```python
rd.Month.unique()
```

```python
sns.barplot(x="Month",y="Estimated Unemployment Rate",hue="Year",palette=["r"],data=rd)
```

```python
sns.set_theme(style="ticks", palette="pastel")
sns.boxplot(x="Month", y='Estimated Employed', palette=["c", "m"], data=rd)
sns.despine(offset=10, trim=True)
```

```python
plt.title('Indian Unemployment')
sns.boxplot(x="Day",y="Estimated Unemployment Rate",hue="Region",data=rd)
plt.show()
```
